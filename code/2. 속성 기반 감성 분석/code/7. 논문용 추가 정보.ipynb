# Characteristics of attribute groups (추가)

```python
df = pd.read_csv('전체_모델속성별_횟수.csv')
```

```python
lst_out = ['proc_yn', 'stor_yn', 'proc_p', 'stor_p', 'proc_n', 'stor_n', 'acc_yn', 'acc_p', 'acc_n']
lst_aspect = ['bat', 'dis', 'cam', 'body', 'func', 'main']

lst_yn = ['bat_yn', 'dis_yn', 'cam_yn', 'body_yn', 'func_yn', 'main_yn']
lst_p = ['bat_p', 'dis_p', 'cam_p', 'body_p', 'func_p', 'main_p']
lst_n = ['bat_n', 'dis_n', 'cam_n', 'body_n', 'func_n', 'main_n']
```

```python
df = df.iloc[:, :-9].drop(lst_out, axis=1)
df = df[df['period']<=3].drop(columns='rel_ord')
```

```python
df
```

```python
# 언급비율
lst_yn_r = []

for col in lst_yn:
    new_col = col + '_r'
    lst_yn_r.append(new_col)
    df[new_col] = df[col] / df['count']    

# 상대적중요도 분모값
df['상대적중요도_분모'] = df[lst_yn_r].sum(axis=1)
lst_상대적중요도 = []

for col in lst_yn_r:
    new_col = '상대적중요도_' + col.replace('_yn_r', '')
    lst_상대적중요도.append(new_col)
    
    df[new_col] = df[col] / df['상대적중요도_분모']
    
# 1년차 상대적중요도 추출
lst_상대적중요도_1p = []

df_1p = df[df['period']==1][['model'] + lst_상대적중요도]

for col in lst_상대적중요도:
    new_col = '1p_' + col
    lst_상대적중요도_1p.append(new_col)
    df_1p = df_1p.rename(columns={col:new_col})
    
    
# 1년차 대비 상대적 중요도
df = pd.merge(df, df_1p, on='model', how='left')

lst_상대적중요도_대비 = []
for col in lst_상대적중요도:
    new_col = '1대비_' + col
    col_1p = '1p_' + col
    lst_상대적중요도_대비.append(new_col)
    df[new_col] = df[col] / df[col_1p]
```

```python
# 긍정비율
lst_p_r = []

for i in range(len(lst_p)):
    new_col = lst_p[i] + '_r'
    lst_p_r.append(new_col)

    df[new_col] = df[lst_p[i]] / df[lst_yn[i]]

# 부정비율
lst_n_r = []

for i in range(len(lst_n)):
    new_col = lst_n[i] + '_r'
    lst_n_r.append(new_col)

    df[new_col] = df[lst_n[i]] / df[lst_yn[i]]

# 긍부정비
lst_pn = []

for i in range(len(lst_yn)):
    new_col = '긍부정_' + lst_aspect[i]
    lst_pn.append(new_col)
    
    pv = lst_p_r[i]
    nv = lst_n_r[i]

    df[new_col] = df[pv] / (df[pv] + df[nv])

# 부긍정비
lst_np = []

for i in range(len(lst_yn)):
    new_col = '부긍정_' + lst_aspect[i]
    lst_np.append(new_col)
    
    pv = lst_p_r[i]
    nv = lst_n_r[i]

    df[new_col] = df[nv] / (df[pv] + df[nv])

# 1년차 대비 부긍정
lst_np_1p = []

df_1p = df[df['period']==1][['model'] + lst_np]

for col in lst_np:
    new_col = '1p_' + col
    lst_np_1p.append(new_col)
    df_1p = df_1p.rename(columns={col:new_col})

df = pd.merge(df, df_1p, on='model', how='left')

lst_np_대비 = []

for col in lst_np:
    new_col = '1대비_' + col
    col_1p = '1p_' + col
    lst_np_대비.append(new_col)
    df[new_col] = df[col] / df[col_1p]
```

```python
df_sel = df.copy()
df_sel = df_sel[['model', 'period'] + lst_상대적중요도 + lst_상대적중요도_대비 + lst_np + lst_np_대비]
```

```python
cols_to_melt = ['_yn', '_yn_r', '_p', '_p_r', '_n', '_n_r']
melted_dfs = []

for col in cols_to_melt:
    cols = [column for column in df.columns if column.endswith(col)]

    melted_df = df.melt(id_vars=['model', 'period'], value_vars=cols, 
                        var_name='variable', value_name=col)

    melted_df['feature'] = melted_df['variable'].apply(lambda x: x.split('_')[0])
    melted_df.drop(columns=['variable'], inplace=True)
    melted_dfs.append(melted_df)

result = melted_dfs[0]
for melted_df in melted_dfs[1:]:
    result = pd.merge(result, melted_df, on=['model', 'period', 'feature'])

```

```python
result = result.rename(columns={'_yn':'출현빈도', '_yn_r':'출현비율', '_p':'긍정빈도', '_p_r':'긍정비율', 
                       '_n':'부정빈도', '_n_r':'부정비율'})
```

```python
df1 = result.set_index(['model', 'period', 'feature']).sort_values(by=['model', 'period', 'feature'])
df2 = result.set_index(['model', 'feature', 'period']).sort_values(by=['model', 'feature', 'period'])
```
