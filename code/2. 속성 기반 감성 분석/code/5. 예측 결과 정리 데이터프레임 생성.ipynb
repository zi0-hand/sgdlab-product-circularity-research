# 1. ATEPC, APC 모델의 감성 예측 결과 결합하기
- 목적    
    : 처음 ATEPC 모델로 속성 단어와 그에 대한 감성을 추출함. 일부 속성에 대한 감성이 예측된 결과가 존재함. 이는 APC 모델로 감성 예측한 결과보다 정확도가 높다고 판단했기 때문에 (수기로 비교한 결과) 동일한 문장에 대해 ATEPC 모델의 감성 예측 결과와 APC 모델의 결과가 다를 경우 ATEPC 모델의 감성 예측 결과를 우선으로 사용함.

```python
df_asp = pd.read_csv('apc_훈련데이터셋.csv')
#df_asp = df_asp.iloc[:100, :]
print('데이터 개수:', df_asp.shape[0])
df_asp.head()
```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

def tokenize_sentence(sentence):
    tokens = word_tokenize(sentence)  
    return tokens

def tokenize_sentence(sentence):
    tokens = word_tokenize(sentence)
    filtered_tokens = [token for token in tokens if not re.match(r'\[[BE]-ASP\]', token)]
    return filtered_tokens

def find_word(tokens_list, target_word):
    indices = [i for i, token in enumerate(tokens_list) if token == target_word]
    return indices[0]
```

```python
# 속성 위치 마킹하기
with tf.device(device_name):
    df_asp['mark_tokens'] = df_asp['mark_sentence'].apply(tokenize_sentence)
    target_word = "B-ASP"  
    df_asp['begin'] = df_asp['mark_tokens'].apply(lambda x: find_word(x, target_word))

    target_word = "E-ASP"  
    df_asp['end'] = df_asp['mark_tokens'].apply(lambda x: find_word(x, target_word))
```

```python
df_asp.head()
```

```python
df_result = pickle.load(open('감성예측결과_데이터프레임.pkl', 'rb'))
df_result = pd.DataFrame(df_result)
print('데이터 개수:', df_result.shape[0])
df_result.head()
```

```python
df_result['sentiment'] = df_result['sentiment'].apply(lambda x: str(x[0]))
df_result = df_result[['text', 'sentiment']]
```

```python
df_con = pd.concat([df_asp, df_result], axis=1)
df_con.head()
```

```python
# apetc에서 도출된 감성과 결합하기 위해서
df_bf = pickle.load(open('감성분석후_속성전처리_데이터프레임.pkl', 'rb'))
print('데이터 개수:', df_bf.shape[0])
df_bf.head(2)
```

```python
lst_tokens = []
lst_position = []
lst_aspect = []
lst_sentiment = []

save_aspect = []
save_tokens = []
save_id = []
save_sentiment = []

save_position = []

with tf.device(device_name):
    for row in tqdm(range(df_bf.shape[0])):
        id = df_bf.loc[row, 'id']
        ori_lst_tokens = list(df_bf.loc[row, 'tokens'])
        lst_position = list(df_bf.loc[row, 'position'])
        lst_aspect = list(df_bf.loc[row, 'aspect'])
        lst_sentiment = list(df_bf.loc[row, 'sentiment'])

        for idx in range(len(lst_aspect)):
            lst_tokens = ori_lst_tokens[:]
            position = lst_position[idx][0]
            aspect = lst_aspect[idx]
            lst_tokens[position] = '[B-ASP]' + aspect + '[E-ASP]'
            mark_sentence = " ".join(lst_tokens)
            sentiment = lst_sentiment[idx]

            save_position.append(position)

            save_tokens.append(mark_sentence)
            save_aspect.append(aspect)
            save_sentiment.append(sentiment)
            save_id.append(id)
```

```python
df_atepc = pd.DataFrame({'id':save_id, 'bf_aspect':save_aspect, 'bf_sentiment':save_sentiment, 
                         'mark_sentence':save_tokens, 'position':save_position})
print('데이터 개수:', df_atepc.shape[0])
df_atepc.head()
```

```python
with tf.device(device_name):
    df_atepc['mark_tokens'] = df_atepc['mark_sentence'].apply(tokenize_sentence)

    target_word = "B-ASP"  
    df_atepc['begin'] = df_atepc['mark_tokens'].apply(lambda x: find_word(x, target_word))

    target_word = "E-ASP"  
    df_atepc['end'] = df_atepc['mark_tokens'].apply(lambda x: find_word(x, target_word))

    df_atepc = df_atepc[['id', 'bf_aspect', 'bf_sentiment', 'begin', 'end']]
```

```python
df_atepc.head()
```

```python
with tf.device(device_name):
    merged_df = df_con.merge(df_atepc, on='id', how='outer')
    merged_df = merged_df[(merged_df['begin_y'] >= merged_df['begin_x']) & (merged_df['end_y'] <= merged_df['end_x'])]
```

```python
with tf.device(device_name):
    df_mer = pd.merge(df_con, merged_df, on=['id', 'mark_sentence', 'ori_feature', 
                                            'feature', 'main_feature', 'sentiment', 'text'], how='left')
    df_mer = df_mer[['id', 'text', 'ori_feature', 'feature', 'main_feature', 
                     'sentiment', 'bf_sentiment', 'bf_aspect']]

df_mer.head()
```

```python
with tf.device(device_name):
    df_info = df_bf[['id', 'model', 'date', 'rating']]
    df_bf_af = pd.merge(df_info, df_mer, on='id')
    print('데이터 개수:', df_bf_af.shape[0])
    df_bf_af.head()
```

# 2. 언어 분류
- 영어권 리뷰만 사용하기 위해서

```python
data = pickle.load(open('감성분석결과_데이터프레임.pkl', 'rb'))
print(data.shape)
data.head(3)
```

```python
import fasttext
model = fasttext.load_model('lid.176.bin')

def detect_language(text):
    text = text.replace('\n', '')
    predictions = model.predict(text)
    predicted_label = predictions[0][0].replace('__label__', '')
    return predicted_label
```

```python
data['lan'] = data['sentence'].apply(detect_language)
```

```python
print('전체 데이터 수:', end=' ')
print(data.shape[0])
print('영어권 데이터 수:', end=' ')
print(data[data['lan'].isin(['en', 'ja', 'ceb'])].shape[0])
print('제외된 데이터 수:', end=' ')
print(data[~data['lan'].isin(['en', 'ja', 'ceb'])].shape[0])
```

```python
# 영어권 리뷰만 추출
data = data[data['lan'].isin(['en', 'ja', 'ceb'])]
data = data.drop('lan', axis=1).reset_index(drop=True)
```

# 3. 데이터프레임 형태 구축

```python
df = data.copy()
print('데이터 개수:', df.shape[0])
df.head(3)
```

```python
def main_feature_sort(lst_sentiment, lst_aspect):
    index = 0
    lst_result = []
    
    for aspect in lst_aspect:
        sentiment = lst_sentiment[index]

        if sentiment=='Negative':
            final_aspect = aspect + '_n'
        elif sentiment=='Neutral':
            final_aspect = aspect + '_m'
        else:
            final_aspect = aspect + '_p'
        lst_result.append(final_aspect)
        index += 1
    
    return list(set(lst_result))
```

```python
df['sort_result'] = df.apply(lambda row: main_feature_sort(row['sentiment'], row['main_feature']), axis=1)
```

```python
labels = ['bat', 'dis', 'cam', 'body', 'func', 'proc', 'stor', 'acc', 'main']
sentiments = ['p', 'm', 'n']

# 새로운 데이터프레임 생성
df_aspect = df.copy()

with tf.device(device_name):
    # 모든 라벨과 감정 조합에 대해 열 생성 및 라벨링
    for label in labels:
        for sentiment in sentiments:
            column_name = f"{label}_{sentiment}"
            df_aspect[column_name] = 0  # 초기값 설정

    # 각 행을 차례로 돌며, 라벨링 진행
    for row in range(df_aspect.shape[0]):
        for aspect in df_aspect.loc[row, 'sort_result']:
            df_aspect.loc[row, aspect] =1
```

```python
df_aspect.head(2)
```

```python
labels = ['bat', 'dis', 'cam', 'body', 'func', 'proc', 'stor', 'acc', 'main']
yesno = ['yn']

# 새로운 데이터프레임 생성
df_aspect_yn = df_aspect.copy()

with tf.device(device_name):
    # 모든 라벨과 감정 조합에 대해 열 생성 및 라벨링
    for label in labels:
        column_name = label +'_yn'
        df_aspect_yn[column_name] = 0  # 초기값 설정

    for row in range(df_aspect_yn.shape[0]):
        for aspect in df_aspect_yn.loc[row, 'sort_result']:
            col_name = aspect.split('_')[0] + '_yn'
            df_aspect_yn.loc[row, col_name] = 1
```

```python
# storage, processor, 기존 mainboard을 모두 포함하는 것을 모두 mainboard로 분류
df_aspect_yn.loc[df_aspect_yn['stor_p']==1, 'main_p'] = 1
df_aspect_yn.loc[df_aspect_yn['stor_m']==1, 'main_m'] = 1
df_aspect_yn.loc[df_aspect_yn['stor_n']==1, 'main_n'] = 1

df_aspect_yn.loc[df_aspect_yn['proc_p']==1, 'main_p'] = 1
df_aspect_yn.loc[df_aspect_yn['proc_m']==1, 'main_m'] = 1
df_aspect_yn.loc[df_aspect_yn['proc_n']==1, 'main_n'] = 1

df_aspect_yn.loc[df_aspect_yn['stor_yn']==1, 'main_yn'] = 1
df_aspect_yn.loc[df_aspect_yn['proc_yn']==1, 'main_yn'] = 1
```

```python
for label in labels:
    for sentiment in sentiments:
        column_name = f"{label}_{sentiment}"
        print(column_name, ':', df_aspect_yn[df_aspect_yn[column_name]==1].shape[0])
    print('')
```

```python
num_sum = 0

for label in labels:
    column_name = label + '_yn'
    aspect_num = df_aspect_yn[df_aspect_yn[column_name]==1].shape[0]
    num_sum += aspect_num
    print(column_name, ':', aspect_num)
    print('')

be_num = num_sum
print('전체 합 :', be_num)
```
