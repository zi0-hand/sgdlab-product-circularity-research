```python
data_path = '/Users/sonjiyeong/Documents/랩실 프로젝트/스마트폰 리뷰마이닝/2. 속성 기반 감성 분석/output/'
```

```python
data = pickle.load(open(data_path + '최종_period_데이터프레임.pkl', 'rb'))
df = data.copy()

df = df[['id', 'model', 'date', 'period', 'sentence']]
df = df[df['period'].isin(['1', '2', '3'])]
```

```python
def tokenize(text):
    return text.split()

def sentence_length(lst):
    return len(lst)

df['sentence_token'] = df['sentence'].apply(tokenize)
df['sentence_length'] = df['sentence_token'].apply(sentence_length)
```

```python
df_mean = df.groupby(['model', 'period']).mean()
df_mean
df_mean.to_excel('단어수_평균.xlsx')
```
