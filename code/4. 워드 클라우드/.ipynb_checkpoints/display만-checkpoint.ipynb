# 1. tf-idf

## 데이터 변환

```python
data = pd.read_csv(path + 'tf-idf용_데이터프레임.csv')
data.head()
data.shape
```

```python
df = data.copy()
#df = df.iloc[:2000, :]
df = df[['id', 'period', 'sentence', 'main_feature', 'f_sentiment']].drop_duplicates()
df.head()
df.shape
```

```python
# 원하는거 문서처럼
lst_asp = ['dis']
lst_col = ['dis_p_1', 'dis_p_2', 'dis_p_3', 'dis_n_1', 'dis_n_2', 'dis_n_3']

df_docu = pd.DataFrame({'feature':lst_col})
df_docu['sentence'] = np.nan
df_docu
```

```python
df_docu.loc[df_docu['feature']=='dis_p_1', 'sentence'] = ' '.join(df[(df['f_sentiment']=='Positive') & 
                                                                     (df['period']==1)]['sentence'].tolist())
df_docu.loc[df_docu['feature']=='dis_p_2', 'sentence'] = ' '.join(df[(df['f_sentiment']=='Positive') & 
                                                                     (df['period']==2)]['sentence'].tolist())
df_docu.loc[df_docu['feature']=='dis_p_3', 'sentence'] = ' '.join(df[(df['f_sentiment']=='Positive') & 
                                                                     (df['period']==3)]['sentence'].tolist())
df_docu.loc[df_docu['feature']=='dis_n_1', 'sentence'] = ' '.join(df[(df['f_sentiment']=='Negative') & 
                                                                     (df['period']==1)]['sentence'].tolist())
df_docu.loc[df_docu['feature']=='dis_n_2', 'sentence'] = ' '.join(df[(df['f_sentiment']=='Negative') & 
                                                                     (df['period']==2)]['sentence'].tolist())
df_docu.loc[df_docu['feature']=='dis_n_3', 'sentence'] = ' '.join(df[(df['f_sentiment']=='Negative') & 
                                                                     (df['period']==3)]['sentence'].tolist())
df_docu
df = df_docu.copy()
```

## 데이터 전처리

```python
stop = stopwords.words('english') + ['tsp', 'tbsp', 'finely','extra', 'chopped' ]
```

```python
def remove_encoding_word(word):
    word = str(word)
    word = word.encode('ASCII', 'ignore').decode('ASCII')
    return word

def remove_encoding_text(text):
    text = str(text)
    text = ' '.join(remove_encoding_word(word) for word in text.split() if word not in stop)
    return text
```

```python
df['sentence'] = df['sentence'].apply(remove_encoding_text)
df
```

## tf-idf 시작

```python
lemma = WordNetLemmatizer().lemmatize
lemma('leaves')
```

```python
def tokenize(document):
    tokens = [lemma(w) for w in document.split() if len(w) > 3 and w.isalpha()]
    return tokens
```

```python
vectorizer = TfidfVectorizer(tokenizer = tokenize, 
                             ngram_range = ((2,2)), 
                             stop_words = stop, 
                             strip_accents = 'unicode')
```

```python
tdm = vectorizer.fit_transform(df['sentence'])
```

```python
data = df.copy()

lst_tfidf = []
lst_dum = []

for asp in lst_asp:
    
    df = data[data['feature'].str.contains(asp)]
    tdm = vectorizer.fit_transform(df['sentence'])

    with tf.device(device_name):
        for i in range(6):
            lst_dum = []
            tfidf_vector = tdm[i]

            non_zero_indices = tfidf_vector.nonzero()[1]
            for idx in tqdm(non_zero_indices):
                word = list(vectorizer.vocabulary_.keys())[list(vectorizer.vocabulary_.values()).index(idx)]
                tfidf_value = tfidf_vector[0, idx]
                lst_dum.append((word, tfidf_value))

            lst_tfidf.append(lst_dum)
```

# 2. 워드 클라우드

```python
with tf.device(device_name):
    w = WordCloud(width=1500, height=1200, mode='RGBA', mask=img,
                  background_color='white', max_words=500).fit_words(dict(lst_tfidf[0]))
    plt.figure(figsize=(15,8))
    _ = plt.imshow(w)
    _ = plt.axis('off')
    _ = plt.title('dis_p_1')
```
